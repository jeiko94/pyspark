```python
import pandas as pd
```

```python
df = pd.read_csv("sales.csv")
df.head(1)
```

```python
filtro = df[df['OrderDate'] > '2019-07-01']
filtro.head(1)
```

**Seleccionar columnas espeficicas**

```python
df_selected = df[['SalesOrderNumber', 'CustomerName', 'Item', 'Quantity', 'UnitPrice']]
df_selected.head(5)
```

**Crear columnas calculadas**

```python
# Crear columna Total y TotalConImpuesto
df['Total'] = df['Quantity'] * df['UnitPrice']
df['TotalConImpuesto'] = df['Total'] + df['TaxAmount']

df[['SalesOrderNumber', 'Total', 'TotalConImpuesto']].head(5)
```

**Renombrar columnas**

```python
# Renombrar columnas
df_renamed = df.rename(columns={
    'SalesOrderNumber': 'NumeroOrden',
    'CustomerName': 'NombreCliente',
    'Item': 'Producto',
    'Quantity': 'Cantidad',
    'UnitPrice': 'PrecioUnitario',
    'Total': 'Total',
    'TotalConImpuesto': 'TotalConImpuesto'
})

df_renamed.head(1)
```

**Limpieza básica (duplicados y faltantes)**

```python
# Revisar duplicados
duplicates = df[df.duplicated(subset=['SalesOrderNumber'], keep=False)]
#duplicates.head()
# Eliminar duplicados manteniendo la primera ocurrencia
df_cleaned = df.drop_duplicates(subset=['SalesOrderNumber'], keep='first')

# Manejar valores faltantes (si hay)
df_cleaned = df_cleaned.dropna() # Eliminar filas con valores faltantes
df_cleaned.head()
```
